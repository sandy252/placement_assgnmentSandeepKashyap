```python
df = df.drop(['id','timestamp','long_summary','apparentTemperatureHighTime','apparentTemperatureLowTime',
                  'apparentTemperatureLowTime','windGustTime','sunriseTime','sunsetTime','uvIndexTime','temperatureMinTime',
                 'temperatureMaxTime','apparentTemperatureMinTime','temperatureLowTime','apparentTemperatureMaxTime'
                 ,'latitude','longitude','product_id'],axis = 1)
print(df.shape)
df.head()
```

```python
df['year'] = df['datetime'].apply(lambda x: x.split('-')[0])
```

```python
df = df.drop(['datetime','year'],axis = 1)
print(df.shape)
df.head()
```

```python
# Reset index before using Onehot Encoder
df = df.reset_index(drop=True)
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_col = df.select_dtypes(include=['object','category']).columns.tolist()
print(cat_col)

#intiate OneHotEncoder and concatinating original rideshare with encoded column rideshare's
for col in cat_col:
    encoder = OneHotEncoder(handle_unknown='ignore')
    enc_rideshare = pd.DataFrame(encoder.fit_transform(df[[col]]).toarray())
    enc_rideshare.columns = encoder.get_feature_names_out([col])
    df = df.drop(col, axis=1)
    df = pd.concat([df, enc_rideshare], axis=1)
```

```python
df.dropna(inplace = True)
X=df.drop(['price'],axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
instances = X.shape[0]
features = X.shape[1]

# Converting each dataframe into a numpy array since each dataframe contains only one column
X = np.array(X).reshape(-1, features)
y = np.array(y).reshape(-1, 1)

identity_vector = np.asarray([[1 for num in range(instances)]])
identity_vector = identity_vector.reshape(-1,1)
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=40)
```

```python
def evaluate_model(test,pred):
    #print(test,pred)
    print("R^2:", r2_score(test, pred))
    print("MAE:", mean_absolute_error(test,pred))
    print("RMSE:",np.sqrt(mean_squared_error(test, pred)))
    print("")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
lr = LinearRegression()
lr.fit(X_train,y_train)
y_val_pred = lr.predict(X_test)
y_train_pred = lr.predict(X_train)

print('Shape:',X_train.shape)
print('Rank:',np.linalg.matrix_rank(X_train))
print('coeff:',lr.coef_)
#Test linear regression model 
print("Training Loss")
evaluate_model(y_train,y_train_pred)
print("Validation Loss")
evaluate_model(y_test,y_val_pred)
```
