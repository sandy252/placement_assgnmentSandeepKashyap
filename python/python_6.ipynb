## Evolutions

```python
evolution_ds = dataset.copy()
```

```python
evolution_ds.shape
```

```python
evolution_ds.columns
```

```python
evolution_ds = evolution_ds[['name','prev_evolution','next_evolution']]
```

```python
evolution_ds
```

```python
evolution_ds['prev_evolution'] = evolution_ds['prev_evolution'].fillna(0)
```

```python
evolution_ds['next_evolution'] = evolution_ds['next_evolution'].fillna(0)
```

```python
evolution_ds['prev_evolution_no'] = evolution_ds['prev_evolution'].apply(lambda x: len(x.split(',')) if isinstance(x, str) else x)
```

```python
evolution_ds['next_evolution_no'] = evolution_ds['next_evolution'].apply(lambda x: len(x.split(',')) if isinstance(x, str) else x)
```

```python
evolution_ds['total_evolutiions'] = evolution_ds['prev_evolution_no'] + evolution_ds['next_evolution_no']
```

```python
evolution_ds.head()
```

```python
eligible_names = evolution_ds[evolution_ds['total_evolutiions'] <= 2]['name']
```

```python
len(eligible_names),eligible_names
```

## spawn time

```python
spawn_ds = dataset.copy()
```

```python
spawn_ds = spawn_ds[['name', 'spawn_time']]
```

```python
spawn_ds.head()
```

```python
spawn_ds['spawn_time'].fillna('0:0', inplace=True)
```

```python
spawn_ds['spawn_time'].isnull().sum()
```

```python
spawn_ds['spawn_time_in_sec'] = spawn_ds['spawn_time'].apply(lambda x: int(x.split(':')[0]) * 60 + int(x.split(':')[1]))
```

```python
spawn_ds.head()
```

```python
spawn_names = spawn_ds[spawn_ds['spawn_time_in_sec'] < 300]['name']
```

```python
len(spawn_names),spawn_names
```
