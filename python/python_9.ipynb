Since there are only few nan values in over all dataframe so here i am going to remove all rows having nan values


```python
def calculate_nan_percentage(dataframe):
    # Calculate the total number of NaN values in each column
    nan_counts = dataframe.isna().sum()

    # Calculate the percentage of NaN values for each column
    nan_percentages = (nan_counts / len(dataframe)) * 100

    return nan_percentages

nan_percentages = calculate_nan_percentage(df)
print(nan_percentages)
```

As we can clearly see the percentage of nan values for each column are less than 1 percent so here I am going to drop all the nan values

```python
df.dropna(inplace=True)
```
